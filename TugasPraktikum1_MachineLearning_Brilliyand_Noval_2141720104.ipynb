## - Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan. (Slice Data)

```python
df = df[['id', 'diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'texture_worst', 'perimeter_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst', 'area_worst']]
df.head()
```

## - Lakukan proses encoding pada kolom "diagnosis".

```python
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
df.head()
```

## - Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
std = StandardScaler()
df['radius_mean'] = std.fit_transform(df[['radius_mean']])
df['texture_mean'] = std.fit_transform(df[['texture_mean']])
df['perimeter_mean'] = std.fit_transform(df[['perimeter_mean']])
df['area_mean'] = std.fit_transform(df[['area_mean']])
df['smoothness_mean'] = std.fit_transform(df[['smoothness_mean']])
df['compactness_mean'] = std.fit_transform(df[['compactness_mean']])
df['concavity_mean'] = std.fit_transform(df[['concavity_mean']])
df['concave points_mean'] = std.fit_transform(df[['concave points_mean']])
df['texture_worst'] = std.fit_transform(df[['texture_worst']])
df['perimeter_worst'] = std.fit_transform(df[['perimeter_worst']])
df['smoothness_worst'] = std.fit_transform(df[['smoothness_worst']])
df['compactness_worst'] = std.fit_transform(df[['compactness_worst']])
df['concavity_worst'] = std.fit_transform(df[['concavity_worst']])
df['concave points_worst'] = std.fit_transform(df[['concave points_worst']])
df['area_worst'] = std.fit_transform(df[['area_worst']])
```

```python
df.head()
```

## - Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
print("Jumlah data latih:", len(X_train))
print("Jumlah data uji:", len(X_test))

print("Proporsi kelas dalam data latih:")
print(y_train.value_counts(normalize=True))

print("Proporsi kelas dalam data uji:")
print(y_test.value_counts(normalize=True))

print("Contoh data latih:")
print(X_train.head())
print(y_train.head())

print("Contoh data uji:")
print(X_test.head())
print(y_test.head())
```
